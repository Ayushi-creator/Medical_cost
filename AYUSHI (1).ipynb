# Numerical vs Numerical

**Scatter**


**scatter plot gives us a representation of where each points in the entire dataset are present with respect to any features(Columns). Scatter plots are available in 2D as well as 3D.**

```python
sns.set_style('darkgrid')
sns.scatterplot(data =df, x='bmi', y='charges')
```

**The scatterplot above suggests that body mass index (BMI) and insurance charges are positively correlated, where customers with higher BMI typically also tend to pay more in insurance costs. (This pattern makes sense, since high BMI is typically associated with higher risk of chronic disease.)To double-check the strength of this relationship,  might like to add a regression line, or the line that best fits the data. We do this by changing the command to sns.regplot**

```python
sns.regplot(data = df, x='bmi', y='charges')
```

```python
sns.scatterplot(df['bmi'], df['charges'], hue=df['smoker'])
plt.title('bmi vs age')
```

**The hue parameter is used for Grouping variable that will produce points with different colors**

**This scatter plot shows that while nonsmokers to tend to pay slightly more with increasing BMI, smokers pay MUCH more.**

**Draw scatter plot age vs charges columns**

```python
sns.scatterplot(df['age'], df['charges'])
```

**we can see if age is increases the charges is also increases.**

```python
sns.scatterplot(df['age'], df['charges'], hue=df['sex'])
plt.show()
```

# Categorical vs Categorical

**Heatmap, clustermap**

**A heat map is a data visualization technique that shows magnitude of a phenomenon as color in two dimensions.**

**in the heatmap parameters are cmap (color map ), center, vmin , vmax, linewidth, annot and many more**

**simple heatmap**

```python
array = np.array([
                  [1, 4, 7], 
                  [34, 76, 56]
                ])


sns.heatmap(array)
```

**correlation**

```python
df.corr()
```

**heatmap with default color palete**

```python
sns.heatmap(df.corr(), cmap='RdBu', linewidth=1, annot=True)
```

**annot_kws	It is one of the optional parameters and if this value is set to true, the data value is written in each cell or if the value is set to false, the data value is not written in each cell.**

**cmap	Mapping of the data values to the color space and it is one of the optional parameters.**

**linewidths	The user uses this parameter to set the width of the line and it is an optional parameter.**

**crosstab**

**crosstab is a table showing the relationship between two or more variables**

---



```python
ctab = pd.crosstab(df['children'], df['sex'])
ctab
```

```python
sns.heatmap(ctab, linewidth = 1)
```

tranpose heatmap

```python
sns.heatmap(ctab.T)
```

**clustermap() method uses a hierarchical clusters to order data by similarity.**

```python
sns.clustermap(pd.crosstab(df['children'], df['sex']), figsize=(6, 6))
```

```python
sns.clustermap(pd.crosstab(df['children'], df['sex']), figsize=(6, 6),annot=True)
```
